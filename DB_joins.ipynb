# JOIN

```python
query_j = "SELECT * from department_dataset join employee on employee.location = department_dataset.location"
mycursor.execute(query_j)
for i in mycursor:
    print(i)
```

# Inner join

```python
query_j = "SELECT department_dataset.id,department_dataset.dept_name , department_dataset.salary from department_dataset left join employee on employee.location = department_dataset.location"
mycursor.execute(query_j)
for i in mycursor:
    print(i)
```

# Right Join

```python
query_j = "SELECT * from department_dataset right join employee on employee.location = department_dataset.location"
mycursor.execute(query_j)
for i in mycursor:
    print(i)
```
